### Working with HDF5 Files
In this notebook we show how to load the North Su HDF5 file and plot a single 30-s trace.

#### Setup

```python
import matplotlib as mpl
mpl.rc("savefig", dpi=300)
mpl.rcParams['axes.linewidth'] = 0.5
mpl.rcParams['font.size'] = 7.5
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

#### Load HDF5 File

```python
import h5py
f = h5py.File('suzette_2015.h5', 'r')
f.items()
```

#### Plot Raw A/D Trace

```python
import numpy as np
import matplotlib.pyplot as plt
gain08_ad = f['gain08_ad'][1,:]
fig = plt.figure(frameon = False, figsize=(7,4.5))
ax = plt.subplot()
pl1 = plt.plot(gain08_ad)
plt.xlim([0, len(gain08_ad)])
ax.set_xlabel('Sample Number')
ax.set_ylabel('A/D Value');
```
